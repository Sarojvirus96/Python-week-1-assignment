Q5. Using a while loop, verify if the number A is purely divisible by number B and if so then how many
times it can be divisible.

Ans.

```python
A = 24
B = 3
count = 0
```

```python
while A % B == 0:
    A /= B
    count += 1

print(f"A is divisible by B {count} times")
```

Q7. What do you understand about mutable and immutable data types? Give examples for both showing
this property.

Ans. In Python, mutable and immutable are two important concepts that describe the behavior of data types.

Immutable data types are those whose values cannot be changed once they are assigned. Any operation on an immutable object actually creates a new object. Examples of immutable data types in Python include:

Ex:- int: Integer values are immutable. When you perform operations on an integer, it creates a new integer object.


```python
x = 5
x += 2  # Creates a new integer object
print(x)
```

Ex:-float: Floating-point numbers are also immutable.


```python
y = 3.14
y *= 2  # Creates a new float object
print(y)
```

Ex:- str: Strings are immutable in Python. Any operation that appears to modify a string actually creates a new string object.
python

```python
s = "Hello"
s += " World"  # Creates a new string object
print(s)
```

##On the other hand, mutable data types are those that allow their values to be changed after they are assigned. Operations on mutable objects modify the object itself, without creating a new object. Examples of mutable data types in Python include:

Ex:- list: Lists are mutable, and you can modify their elements directly.

```python
my_list = [1, 2, 3]
my_list.append(4)  # Modifies the list in-place
print(my_list)
```

Ex:- dict: Dictionaries are mutable and allow you to modify, add, or remove key-value pairs.

```python
my_dict = {"name": "John", "age": 25}
my_dict["age"] = 26  # Modifies the value for the "age" key
print(my_dict)
```

Ex:- set: Sets are mutable and support operations such as adding or removing elements.

```python
my_set = {1, 2, 3}
my_set.add(4)  # Modifies the set in-place
print(my_set)
```
